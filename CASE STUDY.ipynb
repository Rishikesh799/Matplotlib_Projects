# Check null values in our dataset

```python
b=df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.barh(b.index,b.values)
plt.xticks(rotation=90)
plt.ylabel("Column name",fontsize=15)
plt.xlabel("Values",fontsize=15)
plt.show()
```

```python
n=df.isnull().sum()*100/len(df)
```

```python
new_n=n[n>20].index
```

```python
df.drop(columns=new_n,inplace=True)
```
