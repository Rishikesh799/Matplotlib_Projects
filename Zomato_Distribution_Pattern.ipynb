```python
da=new_df.Country.value_counts().nlargest(3)
```

```python
rating=new_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating_count"})
```

```python
rating
```

#Conclusion:
When rating between 4.5 to 4.9 is Excellent.
when rating are between 4.5 to 4.4  is very good.
When rating between 3.5 to 3.9 is good.
When rating between 2.5 to 3.4 is Average.
When rating between 1.8 to 2.4 is poor.

```python
plt.figure(figsize=(15,8))
sns.barplot(x="Aggregate rating",y="Rating_count",data=rating,hue="Rating text")
plt.xlabel("Agregate Rating",fontsize=15,color="r")
plt.ylabel("Rating Count",fontsize=15,color="m")
plt.show()
```

Conclusion: Not rated are high count.
            max no of rating between 2.5 to 3.4.

```python
sns.countplot(x="Rating color",data=rating)
plt.show()
```
