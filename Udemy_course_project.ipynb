# find no of course per subject

```python
k=df["subject"].value_counts()
```

```python
k
```

```python
import matplotlib.pyplot as plt
plt.plot(k.index,k.values,marker="o")
plt.xticks(rotation=60,fontsize=10)
plt.xlabel("Subject",fontsize=15)
plt.ylabel("NO of course",fontsize=15)
plt.title("No of course VS sbject")
plt.show()
```

# how many are paid and free course

```python
l=df["is_paid"].value_counts()
```

```python
l
```

```python
plt.bar(l.index,l.values,width=0.5,color=["b","g"])
plt.xticks(ticks=range(2),labels=["Free","Paid"])
plt.xlabel("Type of course")
plt.ylabel("NO of courses")
plt.title("Total no of course are free and paid")
plt.show()
```
